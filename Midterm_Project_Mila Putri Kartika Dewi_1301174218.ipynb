### Korelasi antar variabel

```python
# Memproses data dengan menggunakan library "corr" untuk mendapat hasil korelasi antar variabel.

corelation_1 = data_diabetes.corr(method = 'pearson')
corelation_1
```

```python
# Visualisasi data dari korelasi antar semua variabel. 

f,ax = plt.subplots(figsize=(20,20))
sns.heatmap(corelation_1, annot=True, linewidths=.5, fmt='.3f', ax=ax)
plt.title('Korelasi Antar Variabel',y=1,size=20)
plt.show
```

### Korelasi antara variabel dan target

```python
# Hasil korelasi antar variabel target

col = data_diabetes.corrwith(data_diabetes["Outcome"], method = 'pearson')
col
```

```python
k= 10
corelation_2 = corelation_1.nlargest(k,'Outcome')['Outcome'].index
print(corelation_2)
hasil = np.corrcoef(data_diabetes[corelation_2].values.T)
f , ax = plt.subplots(figsize = (14,12))
sns.heatmap(hasil, vmax=.8, linewidths=0.01,square=True,annot=True,cmap='viridis',
            linecolor="white",xticklabels = corelation_2.values ,annot_kws = {'size':15},yticklabels = corelation_2.values)
plt.title('Korelasi Antar Varabel Target',y=1,size=16)
```

#### Penjelasan

Korelasi adalah teknik perhitungan statustika yang  dilakukan untuk mengetahui hubungan keterkaitan antar variable dalam sebuah dataset. Sebelum melakukan proses korelasi dilakukan uji normalisasi untuk mengetahui data tersebut berdistribusi normal atau tidak, karena syarat menggunakan korelasi person data harus berdistribusi normal. ### Soal 3 (30 poin)
Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.

```python
# Membaca dataset kedalam dataframe pandas 

df_HangSeng= pd.read_csv("Hang_Seng.csv") 
df_Nasdaq = pd.read_csv("Nasdaq.csv")
df_Nikkei = pd.read_csv("Nikkei.csv")
```

```python
# Missing value pada dataset HangSeng.csv 

df_HangSeng.isna().sum()
```

```python
df_HangSeng.info()
```

```python
# Missing value pada dataset Nikkei.csv 

df_Nikkei.isna().sum()
```

```python
# Merubah type atribut "Date" ke datetime

df_HangSeng['Date'] = pd.to_datetime(df_HangSeng['Date'])
df_Nasdaq['Date'] = pd.to_datetime(df_Nasdaq['Date'])
df_Nikkei['Date'] = pd.to_datetime(df_Nikkei['Date'])
```

```python
# Filtering data untuk tahun 2019

HangSeng = df_HangSeng[(df_HangSeng['Date'] >= '2019-01-02') & (df_HangSeng['Date'] <= '2019-12-31')]
Nasdaq = df_Nasdaq[(df_Nasdaq['Date'] >= '2019-01-02') & (df_Nasdaq['Date'] <= '2019-12-31')]
Nikkei = df_Nikkei[(df_Nikkei['Date'] >= '2019-01-02') & (df_Nikkei['Date'] <= '2019-12-31')]
Nikkei.head()
```

### Preprocessing Data 

```python
# Menangani missing value dengan menggunakan library "SimpleImputer"

from sklearn.impute import SimpleImputer

#dataset Hang Seng 
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(HangSeng[['Open','High','Low','Close','Adj Close','Volume']])
HangSeng[['Open','High','Low','Close','Adj Close','Volume']] = imputer.transform(HangSeng[['Open','High','Low','Close','Adj Close','Volume']])

#dataset Nikkei
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(Nikkei[['Open','High','Low','Close','Adj Close','Volume']])
Nikkei[['Open','High','Low','Close','Adj Close','Volume']] = imputer.transform(Nikkei[['Open','High','Low','Close','Adj Close','Volume']])
```

```python
# Melakukan perhitungan untuk mendapatkan nilai daily percentage change 

HangSeng["Daily Percentage Change"] = HangSeng["Adj Close"].pct_change()*100
Nasdaq["Daily Percentage Change"] = Nasdaq["Adj Close"].pct_change()*100
Nikkei["Daily Percentage Change"] = Nikkei["Adj Close"].pct_change()*100
```

```python
# Melakukan drop untuk variabel yang terdapat nilai "NaN" atau kosong.

HangSeng.dropna(axis = 0, inplace = True)
Nasdaq.dropna(axis = 0, inplace = True)
Nikkei.dropna(axis = 0, inplace = True)
```

### Dataset Hang Seng

```python
# klasifikasi dari nilai daily percentage change.

def Trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"

HangSeng["Trend"]= HangSeng["Daily Percentage Change"].apply(lambda x:Trend(x))
HangSeng.head()
```

```python
# Tampilan jumlah dari masing-masing trend analysis

HangSeng_data = HangSeng.groupby(['Trend']).count()
print(type(HangSeng.groupby('Trend', axis=0)))
HangSeng_data['Daily Percentage Change']
```

```python
# Visualisasi dataset "Hang Seng" dari trend analysis 

colors_list = ['gold', 'Chocolate','lightgrey', 'lightskyblue', 'lightgreen', 'pink','lightgrey']
pie_label = sorted([i for i in HangSeng.loc[:, 'Trend'].unique()])
explode_list = [0.1, 0, 0, 0, 0, 0] # ratio for each continent with which to offset each wedge.
HangSeng_data['Daily Percentage Change'].plot(kind='pie',
                                            figsize=(15, 8),
                                            autopct='%1.1f%%', 
                                            startangle=310,
                                            labels=None,
                                            shadow=True,          
                                            pctdistance=1.5,   
                                            colors=colors_list,  
                                            explode=explode_list 
                                            )
 
plt.axis('equal') 
plt.legend(labels=pie_label, loc='upper left') 

plt.show()
```

Sedangkan untuk presentase tingkat penurunan dari perubahan harga saham (Daily Percentage Change) yang paling rendah berada pada tingkat “Among top losers” dengan presentase sebesar 0,4%, kemudian pada tingkat “Negative ” dengan presentase sebesar 97,1%, kemudian pada tingkat “Slight Negative” dengan presentase sebesar 14,6% dan tingkat yang paling rendah berada pada “Slight or no Change” menunjukan presentase sebesar 52,9%. 


### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
df_combine = pd.concat([HangSeng['Daily Percentage Change'], Nasdaq['Daily Percentage Change'], Nikkei['Daily Percentage Change']], axis=1)
df_combine.columns = ['HangSeng','Nasdaq','Nikkei']
volatility = df_combine.rolling(7).std()*np.sqrt(7)
volatility.plot(figsize = (15, 7))
```

### Penjelasan

Volatilitas merupakah hal yang terpenting dalam perhitungan saham, ini gunakan untuk mengetahui besarnya jarak antar naik dan turunnya harga saham disuatu perusahaan. Dalam analisis ini akan dilakukan perhitungan volatilitas dengan rentan waktu 1 minggu atau  7 hari. Data tersebut menunjukan bahwa volatilitas tertinggi diperoleh dari saham perusahaan “Nasdaq”, berikutnya volatilitas menengah diperoleh dari saham perusahaan “Nikkei” dan yang terakhir dari perusahaan “HangSeng”. 

Dari analisis data diatas diperoleh suatu informasi bahwa saham yang memiliki volatilitas tinggi berarti kenaikan harga yang cukup tinggi namun dapat turun dengan cepat juga, sehingga saham yang memiliki volatilitas tinggi cenderung mendapatkan keuntungan yang lebih tinggi dan juga berpotensi mendapatkan keuntungan rendah. Sedangkan untuk saham yang memiliki volatilitas rendah lebih mengarah ke harga yang stabil, dengan tingkat resiko yang rendah juga.
